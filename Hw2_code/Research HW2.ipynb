```python
from scipy.stats import geom
import matplotlib.pyplot as plt
#
# X = Discrete random variable representing number of throws
# p = Probability of the perfect throw
#
X = [i for i in range(100)]
# p = 1/9 # 3 item in 3 slots
# p = 1/36 # 6 item in 3 slots
p = 1/81 # 3 item in 5 slots
#
# Calculate geometric probability distribution
#
geom_pd = geom.pmf(X, p)
#
# Plot the probability distribution
#
fig, ax = plt.subplots(1, 1, figsize=(20, 8))
ax.plot(X, geom_pd, 'bo', ms=8, label='geom pmf')
plt.ylabel("Probability", fontsize="18")
plt.xlabel("X - No. of Times", fontsize="18")
plt.title("Geometric Distribution - No. of Times Vs Probability", fontsize="18")
ax.vlines(X, 0, geom_pd, colors='b', lw=5, alpha=0.5)
```

```python
import numpy as np
from itertools import combinations_with_replacement
num_throws = 100
outcomes = []
for i in range(num_throws):
  # let's roll the die
  # comb = list(combinations_with_replacement(["CHERRY", "LEMON", "ORANGE"],3))
  outcome = np.random.choice(["CHERRY", "LEMON", "ORANGE"],3)
  outcomes.append(outcome)


val, cnt = np.unique(outcomes, return_counts=True)
prop = cnt / len(outcomes)

# Now that we have rolled our die 10000 times, let's plot the results
plt.bar(val, prop)
plt.ylabel("Probability")
plt.xlabel("Outcome")
plt.show()
plt.close()
```

```python
import math

def poisson_distribution(k, lambd):
    return (lambd ** k * np.exp(-lambd)) / np.math.factorial(k)
```

```python
from scipy.stats import binom, poisson
import numpy as np
import matplotlib.pyplot as plt
n1 = 20
n2 = 60
n3 = 3600
λ = 9

x = np.arange(20)
print('x : ',x)
X2 = binom.rvs(n=n2, p=λ/n2, size=100000)
X3 = binom.rvs(n=n3, p=λ/n3, size=100000)

Z = poisson.rvs(9, size=100000)
X1 = binom.rvs(n=n1, p=λ/n1, size=100000)
print(f'P(X1<=9) of a Binomial RV with n={n1}: ' + str(X1[X1<9].shape[0]/X1.shape[0]))
print(f'P(X2<=9) of a Binomial RV with n={n2}: ' + str(X2[X2<9].shape[0]/X2.shape[0]))
print(f'P(X3<=9) of a Binomial RV with n={n3}: ' + str(X3[X3<9].shape[0]/X3.shape[0]))
print(f'P(Z<=9) of a Poisson RV with λ={λ}: ' + str(Z[Z<9].shape[0]/Z.shape[0]))
```
